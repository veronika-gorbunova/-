```python
import sys
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
from utility import util
from matplotlib import colors
from matplotlib.colors import hsv_to_rgb
```

```python
image  = cv.imread('Togtik.jpg')
image  = cv.cvtColor(image , cv.COLOR_BGR2RGB)
plt.imshow(image )
```

```python
rgb_image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
hsv_image = cv.cvtColor(image, cv.COLOR_BGR2HSV)
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

image = gray_image

channels = [0]
histSize = [256]
range = [0, 256]

gs = plt.GridSpec(1, 2)
plt.figure(figsize=(10, 8))
plt.subplot(gs[0])
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
edged = cv.Canny(gray_image, 10, 250)
plt.imshow(edged, cmap='gray')
```

```python
kernel = cv.getStructuringElement(cv.MORPH_RECT, (15, 15))
closed = cv.morphologyEx(edged, cv.MORPH_CLOSE, kernel)
plt.imshow(closed, cmap='gray')
cv.imwrite("closed.jpg", closed)
```

```python
(cnt, hierarchy) = cv.findContours( 
    closed.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE) 
rgb = cv.cvtColor(closed, cv.COLOR_BGR2RGB) 
cv.drawContours(rgb, cnt, -1, (0, 255, 0), 2) 
  
plt.imshow(rgb) 
```
